```python
#importing required libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('position_salaries.csv')  
```

```python
X=dataset.iloc[:,1:2].values # Will select all rows & 'level' column. 
y=dataset.iloc[:,2].values  # Will select all rows & 'salary' column. 
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest,ytrain, ytest=train_test_split(X, y, test_size=0.2, random_state=0)
from sklearn.linear_model import LinearRegression
lin=LinearRegression() #LinearRegression() will create the model.
lin.fit(X,y)
```

```python
def viz_linear():
    plt.scatter(X,y , color='Red') #Actual datapoints based on dataset
    plt.plot(X,lin.predict(X) , color='blue') # Plot the line of predictions from the model
    plt.title('Linear Regression Model')
    plt.xlabel('Position Level')
    plt.ylabel('Salary')
    plt.show()
    return()

viz_linear()
```

Observation: Data points are still scatter & not fit to the line. In order to fit the datapoints of scatter plot we will use Polynomial Regression.

```python
#Polynomial Regression = Linear Regression on Polynomial Features
from sklearn.preprocessing import PolynomialFeatures
polf=PolynomialFeatures(degree=4)
x_pol=polf.fit_transform(X)
pol=LinearRegression() # fitting PolynomialFeatures at degree 4
pol.fit(x_pol,y)
```

```python
def viz_Polynomial():
    plt.scatter(X, y , color='red')
    plt.plot(X, pol.predict(polf.fit_transform(X)),color='blue')
    plt.title('Linear Regression with Polynomial with degree 4')
    plt.xlabel('Position Level')
    plt.ylabel('Salary')
    plt.show()
    

viz_Polynomial()
```

```python
lin.predict([[5.5]]) # Predict the salary of the 5.5 level of Person using Linear model.
```
